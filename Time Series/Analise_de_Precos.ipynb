# Trabalho de Séries Temporais
Análise de preços de drogas não tradicionais.

```r
PATH <- 'https://raw.githubusercontent.com/juan1411/Statistics-Degree-2022'
PATH <- paste0(PATH, '/main/Time%20Series/precos.csv')
df <- read.csv(PATH, header=T)
head(df)
```

```r
unique(df$sub_region)
```

```r
unique(df$country)
```

```r
unique(df$drug)
```

```r
coca <- df[df$drug == 'Cocaine',]
hero <- df[df$drug == 'Heroin',]

plot(coca[coca$country=='United States of America', 4],
     coca[coca$country=='United States of America', 6],
     main='Cocaine in United States of America', 
     xlab='Year', ylab='Retail Price', type='l', lwd=2)
points(coca[coca$country=='United States of America', 4],
       coca[coca$country=='United States of America', 6],
       pch=19)
```

```r
acf(coca[coca$country=='United States of America', 6],
    main='Series - Retail Price in USA', lwd=2)
```

```r
pacf(coca[coca$country=='United States of America', 6],
     main='Series - Retail Price in USA', lwd=2)
```
